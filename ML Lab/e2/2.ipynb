```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
from sklearn.metrics import r2_score
```

```python
data = {
    'Car_Age': [4, 4, 5, 5, 7, 7, 8, 9, 10],
    'Price': [6300, 5800, 5700, 4500, 4500, 4200, 4100, 3100, 2100]
}

df = pd.DataFrame(data)
```

```python
x = df['Car_Age'].values.reshape(-1, 1)
y = df['Price'].values
model = LinearRegression()
model.fit(x, y)
```

```python
predictions = model.predict(x);
```

```python
%matplotlib inline
plt.scatter(x, y, marker='+', color='red', label='Actual Price')
plt.plot(x, predictions, color='blue', label='Predicted Price')
plt.xlabel("Car Age (in years)")
plt.ylabel("Price (in dollars)")
plt.legend()
```

```python
print(f"y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}")
print(f'Coefficient: {model.coef_[0]:.2f}') # if the coefficient is negative then negative correlation. we can even infer from the plot.
```

```python
print(f"R2 Score: {r2_score(y, predictions):.2f}")  # if nearer to 1 then we have the perfect fit model.
```

```python
test_car_age = np.array([2.5, 4, 8.6, 4.5]).reshape(-1, 1);
predictions = model.predict(test_car_age)

for car_age, price in zip(test_car_age, predictions):
    print(f"For a {car_age[0]} years old car, price is ${price:.2f}.")
```
